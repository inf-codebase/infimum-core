## 3. Custom Tools

Create custom tools by inheriting from the `Tool` base class.

```python
# Example: Create a custom tool
from core.ai.llm import Tool

class ReverseStringTool(Tool):
    """Custom tool to reverse a string"""
    def __init__(self):
        super().__init__(
            name="reverse_string",
            description="Reverses a given string"
        )
    
    def execute(self, input_text):
        """Execute the reverse string operation"""
        return input_text[::-1]

# Create agent with custom tool
custom_tool = ReverseStringTool()
agent = Agent(tools=[custom_tool])

# Test the tool directly
result = custom_tool.execute("Hello World")
print(f"Reversed string: {result}")
print(f"Agent has tool: {custom_tool.name in agent.tool_manager.get_available_tools()}")
```

## 5. Memory System

The `Memory` class manages short-term and long-term memory for the agent.

```python
# Example: Using Memory directly
memory = Memory(max_short_term=5)

# Add to short-term memory
memory.add_to_short_term({"role": "user", "content": "What is Python?"})
memory.add_to_short_term({"role": "assistant", "content": "Python is a programming language."})

# Get recent entries
recent = memory.get_recent_entries()
print(f"Recent entries: {len(recent)}")
for entry in recent:
    print(f"  {entry['role']}: {entry['content']}")
```
