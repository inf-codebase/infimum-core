## 1. Speech-to-Text

Transcribe audio files to text using various providers (Whisper, HuggingFace, etc.).

```python
# Example 1: Initialize Speech2Text with Whisper using model name
# Whisper model names: tiny, base, small, medium, large, large-v2, large-v3
stt = Speech2Text(provider="whisper", model_name="small")

print(f"Speech2Text initialized with provider: whisper")
print(f"Model: small")
print(f"Provider: {stt._provider.__class__.__name__}")
```

```python
# Example 2: Transcribe an audio file
# Note: Make sure you have an audio file (mp3, wav, etc.)
audio_file = "test_audio.wav"
 
# # Transcribe with automatic language detection
transcript = stt.transcribe(audio_file)
print(f"Transcript: {transcript}")
 
# # Transcribe with specified language (faster, more accurate)
transcript_en = stt.transcribe(audio_file, language="en")
print(f"Transcript (English): {transcript_en}")

print("💡 Transcription example shown (commented out - requires audio file)")
print("   Usage: stt = Speech2Text(provider='whisper', model_name='small')")
print("   Then: transcript = stt.transcribe('your_audio_file.mp3')")
```

## Whisper Model Options

Whisper supports different model sizes:
- **tiny**: Fastest, least accurate (~39M parameters)
- **base**: Good balance (~74M parameters)
- **small**: Better accuracy (~244M parameters) - Recommended
- **medium**: High accuracy (~769M parameters)
- **large**: Best accuracy (~1550M parameters)
- **large-v2**, **large-v3**: Latest large models

```python
# Example: Using different Whisper models

# Small model (recommended for most use cases)
stt_small = Speech2Text(provider="whisper", model_name="small")
print("✅ Small model initialized")

# Base model (faster, less accurate)
stt_base = Speech2Text(provider="whisper", model_name="base")
print("✅ Base model initialized")

# Large model (best accuracy, slower)
# stt_large = Speech2Text(provider="whisper", model_name="large")
# print("✅ Large model initialized")

# Using custom model path
# stt_custom = Speech2Text(provider="whisper", model_path="/path/to/custom/model.pt")
# print("✅ Custom model initialized")
```

## Summary

**Speech2Text**:
- Transcribe audio files to text
- Supports multiple providers (Whisper, HuggingFace)
- Automatic language detection or manual specification
- Batch processing support
- Handles file paths, bytes, or streams

**Text2Speech**:
- Synthesize speech from text
- Supports multiple providers (Coqui, HuggingFace)
- Multi-speaker support
- Save to file or return bytes

**Key Features**:
- Facade pattern hides complexity
- Automatic preprocessing
- Provider abstraction
- Easy to switch providers
